```python
import matplotlib.pyplot as plt
import numpy as np

xaxis=np.array(["c","c++","java","python"])
yaxis=np.array([45,65,88,93])
X_axis = np.arange(len(xaxis))
z=np.array([30,40,50,60])
plt.xlabel("Programmings",fontsize=15)
plt.ylabel("Rates")
plt.bar(X_axis,yaxis,color='r',width=0.2,edgecolor='b',linewidth=0,label="shivay")
plt.bar(X_axis+0.20,z,color='g',width=0.2,edgecolor='r',linewidth=0,label="shivay2")
plt.legend()
plt.title("Testing123")
plt.show()
```
