```python
import numpy as np
import matplotlib.pyplot as plt

```

```python

def graeint(x,y):
    # m=slop b= cpfiecent,
    m_curr=b_crr=0
    itre=10000
    n=len(x)
    plt.scatter(x,y,color='red',marker="+",linewidths=5)
    plt.show()
    learning_rate=0.08
    for i in range(itre):
        # line formula 
        # y=mx+b 
        yp=m_curr*x+b_crr
        
        plt.plot(x,yp,color="green")

        # summation function of mean sqaure error 
        cost=1/n * sum([val**2 for val in (y-yp)])

        # partial derivation 
        md = -(2/n)*sum(x*(y-yp))
        bd = -(2/n)*sum(y-yp)

        m_curr=m_curr-learning_rate *md
        b_crr=b_crr-learning_rate *bd
#         print(f'[m={m_curr}] ,   [b={b_crr}]      [cost={cost}]')
        
```

```python

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

```

```python
graeint(x,y)
```
